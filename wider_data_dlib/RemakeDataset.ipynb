```python
class CreateDataset:
  def __init__(self):
    self.pic_names = []
    self.dir_name = 'our_new_dataset/'
    self.pic_dir = {}

  def get_pic_names(self, folder_name):
    self.pic_names = listdir(folder_name)
    
  def get_face_bbx(self):
    file_name = 'wider_face_val_bbx_gt.txt'
    f = open('our_dataset/' + file_name, 'r')
    g = open(self.dir_name + file_name, 'w')

    write_mode = False
    index = 1
    for line in f.readlines():
      if line[-4:-1] == 'jpg':
        dir_pos = line.find('/')
        line = line[(dir_pos+1):-1]
        if line in self.pic_names:
          write_mode = True

          new_name = str(index) + '.jpg'
          self.pic_dir[line] = new_name
          g.write(new_name + '\n')

          index+=1
        else:
          write_mode = False
      else:
        if write_mode:
          g.write(line)    

    f.close()
    g.close()
  
  def rename_imgs(self):
     for pic_name in self.pic_names:
       if pic_name[-3:] == 'jpg':
          original = r'our_dataset/' + pic_name
          target = r''+self.dir_name + 'pics/' + self.pic_dir[pic_name]
          shutil.copyfile(original, target)
```

```python
dataset = CreateDataset()
dataset.get_pic_names('our_dataset')
dataset.pic_names
```
